```python
import pandas as pd

df=pd.read_csv("student-mat.csv")

df.sample(10)
```

```python
df["alc"]=df["Walc"]+df["Dalc"]
```

```python
df.describe()
```

# trinken Landkinder mehr?

```python
rural=df["address"]=='R'
df_rural_alc=df.loc[rural, ["Dalc", "Walc"]]
df_rural_alc["Dalc"].mean()+df_rural_alc["Walc"].mean()
```

```python
urban=df["address"]=='U'
df_urban_alc=df.loc[urban, ["Dalc", "Walc"]]
df_urban_alc["Dalc"].mean()+df_urban_alc["Walc"].mean()
```

# Haben Studierende, die häufig krank sind (health) häufigere Fehlzeiten (absences)

```python
low_health=df["health"]<3
low_health_absence=df.loc[low_health, "absences"].mean()
print("absence of students with health problems (health <= 2):", low_health_absence)
```

```python
print("absence of students with health >= 3:", df.loc[df["health"]>=3, "absences"].mean())
```

# Welche Merkmale unterscheiden sich besonders stark zwischen Studierenden unterschiedlichen Geschlechts? (sex)

```python
df_male=df[df["sex"]=='M']
df_female=df[df["sex"]=='F']
```

```python
df_male.describe()
```

```python
df_female.describe()
```
